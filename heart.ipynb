```python
import matplotlib.pyplot as plt
import numpy as np

# Create the angles for the points on the heart shape
angles = np.linspace(0, 2*np.pi, 400)

# Create the x and y coordinates for the points on the heart shape
x = 16 * np.sin(angles)**3
y = 13 * np.cos(angles) - 5 * np.cos(2*angles) - 2 * np.cos(3*angles) - np.cos(4*angles)

# Plot the heart shape
plt.plot(x, y)

# Add some padding to the figure
plt.xlim(-20, 20)
plt.ylim(-20, 20)

# Show the heart shape
plt.show()
```
